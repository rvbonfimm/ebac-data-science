# 1 - Traga a média e a mediana dos preços - coluna Preco_Normal - por categoria de produto.

```python
# Seu código aqui para a média
df.groupby('Categoria')['Preco_Normal'].mean().reset_index().sort_values(by='Preco_Normal', ascending=True)
```

### Outliers

Sim, podemos identificar uma série de outliers presente nos dados (exatos 43 registros dentre os 447 da categoria "lacteos"). Para identificá-lo, basta olharmos para o valor máximo definido pelo boxplot e filtrar os dados pela coluna "Preco_Normal" no dataframe que passar disso.

Dica: Para trazer apenas os dados da categoria que você deseja você pode usar o df.loc[df['Categoria'] == 'CATEGORIA ESCOLHIDA'

```python
# Seu código aqui

filtered_data = df.loc[df['Categoria'] == 'lacteos']
fig = px.box(filtered_data, y='Preco_Normal',
             title='Distribuição dos Preços dos produtos - Categoria Lacteos',
             points='all',
             labels={'Preco_Normal': "Preço Normal"},
             notched=True
            )

fig.show()
```

# 4 - Plote um gráfico de barras onde temos a média de descontos por categoria.

```python
# Seu código aqui

media_descontos_por_categoria = df.groupby('Categoria')['Desconto'].mean().reset_index().sort_values(by='Desconto', ascending=True)
fig = px.bar(media_descontos_por_categoria,
             x='Desconto',
             y='Categoria',
             title='Média de descontos por categoria',
             labels={'Desconto': "Média de Descontos", 'Categoria': "Categoria"},
             color='Categoria',
             text='Desconto'
            )

fig.update_traces(texttemplate='%{value:0.2f}')
fig.show()
```

# 5 - Plote um gráfico de mapa interativo agrupando os dados por categoria, marca e trazendo a média de desconto.

```python
# Seu código aqui
data = df.groupby(['Categoria', 'Marca'])['Desconto'].mean().reset_index()
fig = px.treemap(data,
                 path=['Categoria', 'Marca'],
                 values='Desconto',
                 color='Categoria'
                )
fig.update_traces(root_color="lightgrey")
fig.update_layout(margin = dict(t=50, l=50, r=50, b=50))
fig.update_traces(marker=dict(cornerradius=5))
fig.show()
```
